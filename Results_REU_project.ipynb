```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt


drive.mount('/content/drive')

cnn_results_path = '/content/drive/My Drive/CNN_model_predictions.csv'
lstm_results_path = '/content/drive/My Drive/LSTM_model_predictions.csv'
transformer_results_path = '/content/drive/My Drive/Transformer_model_predictions.csv'


cnn_df = pd.read_csv(cnn_results_path)
lstm_df = pd.read_csv(lstm_results_path)
transformer_df = pd.read_csv(transformer_results_path)


plt.figure(figsize=(14, 7))
plt.plot(cnn_df['original_values'], label='Original Values', color='black')

plt.plot(cnn_df['CNN_predicted_values'], label='CNN Predictions', color='blue')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('CNN Model Predictions vs Original Values')
plt.legend()
plt.show()


plt.figure(figsize=(14, 7))
plt.plot(lstm_df['original_values'], label='Original Values', color='black')
plt.plot(lstm_df['LSTM_predicted_values'], label='LSTM Predictions', color='red')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('LSTM Model Predictions vs Original Values')
plt.legend()
plt.show()


plt.figure(figsize=(14, 7))
plt.plot(transformer_df['original_values_Transformer'], label='Original Values', color='black')
plt.plot(transformer_df['predicted_values_transformer'], label='Transformer Predictions', color='green')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Transformer Model Predictions vs Original Values')
plt.legend()
plt.show()
```
